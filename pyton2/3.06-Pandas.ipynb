* Crie um dataframe `gols` com duas colunas `pa√≠s` e `total_gols` com o total de gols marcados por cada pa√≠s durante todas as copas do mundo, juntando gols em casa (`gols_1`) e gols como time visitante (`gols_2`).

Segue um exemplo ilustrativo com o formato do dataframe resultado:


````python
# Criando o dataframe de gols com o total de gols marcados por cada pa√≠s
gols = pd.concat([
    wc.groupby('time_1')['gols_1'].sum(),  # Gols marcados como time da casa
    wc.groupby('time_2')['gols_2'].sum()   # Gols marcados como time visitante
], axis=1)

# Renomeando as colunas para facilitar
gols.columns = ['gols_casa', 'gols_visitante']

# Somando os gols de casa e visitante
gols['total_gols'] = gols['gols_casa'] + gols['gols_visitante']

# Selecionando a coluna total_gols e ordenando de forma decrescente
gols = gols[['total_gols']].sort_values(by='total_gols', ascending=False)

# Resetando o √≠ndice para exibir como pa√≠s
gols.reset_index(inplace=True)

# Exibindo o dataframe
print(gols.head())
````

#### Q5. 
Qual pa√≠s tomou mais cart√µes amarelos somando todas as copas?

Neste exerc√≠cio voc√™ vai trabalhar com as colunas `cartao_amarelo_1` e `cartao_amarelo_2` onde cada observa√ß√£o √© uma string represetando uma lista dos cart√µes amarelos de um √∫nico jogo na forma `minuto|placar|jogador(a)`. Por exemplo:
```
['16‚Äô|1:0|Rosana G√≥mez', '20‚Äô|2:0|Gabriela Ch√°vez]
```

Recomendo criar colunas `num_cartoes_1` e `num_cartoes_2` a partir de cada coluna `cartao_amarelo_X` usando o m√©todo gen√©rico `apply` para chamar uma fun√ß√£o que voc√™ vai criar para transformar uma observa√ß√£o de cart√£o amarelo em uma contagem de elementos da lista. 

Lembre de levar em considera√ß√£o que muitas observa√ß√µes s√£o `NaN`. 

Em seguida fa√ßa sua m√°gica para agrupar as informa√ß√µes por pa√≠s, acumular os cart√µes de jogos em casa e jogos visitante, e produzir o resultado final como apresentado a seguir (valores inventados):

```
pa√≠s        cart√µes amarelos
Argentina   97
England     93
Australia   93
...
```

````python
# Fun√ß√£o para contar o n√∫mero de cart√µes amarelos em uma lista de cart√µes
def contar_cartoes(cartao_coluna):
    # Se a c√©lula for NaN, retorna 0
    if isinstance(cartao_coluna, str):
        # Divida a string pelos separadores de cart√£o e conte os itens (cart√µes)
        return len(cartao_coluna.split(","))
    else:
        return 0

# Criando novas colunas para o n√∫mero de cart√µes amarelos
wc['num_cartoes_1'] = wc['cartao_amarelo_1'].apply(contar_cartoes)
wc['num_cartoes_2'] = wc['cartao_amarelo_2'].apply(contar_cartoes)

# Agora somamos os cart√µes amarelos para os times da casa e visitante
# Agrupar por time_1 e time_2
cartoes_por_pais = wc.groupby('time_1')['num_cartoes_1'].sum() + wc.groupby('time_2')['num_cartoes_2'].sum()

# Organizando o resultado final em ordem decrescente
cartoes_por_pais = cartoes_por_pais.sort_values(ascending=False)

# Exibindo o resultado final
print(cartoes_por_pais.head())
````

#### Q6.
Qual o top10 jogadores com mais gols em copa? Considere gols em jogo e gols de p√™nalti.

Para conseguir essa informa√ß√£o, voc√™ precisar√° trabalhar com as colunas:
```
10  gols_1_detalhes         
11  gols_2_detalhes    
14  gols_1_penalti     
15  gols_2_penalti     
```

Essas tamb√©m s√£o colunas textuais, onde cada observa√ß√µes apresenta todos os gols de uma partida separados pelo caracter `|`. Cada gol est√° na forma `'Jogador(a) ¬∑ minuto‚Äô'`. Por exemplo:
```
'Alex Morgan ¬∑ 12‚Äô|Rose Lavelle ¬∑ 20‚Äô'
```

Lembre de levar em considera√ß√£o que muitas observa√ß√µes s√£o `NaN`. 

Recomendo criar um dicion√°rio √† parte, onde cada chave ser√° um jogador encontrado nessas colunas do dataframe `wc` e o valor correspondente ser√° a contagem de ocorr√™ncias desses nomes.

Em seguida basta converter o seu dicion√°rio em um dataframe e imprim√≠-lo na forma (valores n√£o s√£o inventados üòÅ):
```
jogador(a)      num_gols 
Marta           17
Miroslav Klose  16
... 
```

````python
import pandas as pd
from collections import Counter

# Supondo que o DataFrame wc j√° esteja carregado e processado conforme o exerc√≠cio
# Fun√ß√£o para contar os gols dos jogadores em cada partida (em casa e fora)

def contar_gols(coluna):
    # Vamos usar a fun√ß√£o `apply` para extrair o nome dos jogadores e contar os gols
    gols = []
    for linha in coluna.dropna():  # Remove valores NaN
        jogadores = linha.split('|')  # Divide os gols por '|'
        for jogador in jogadores:
            nome_jogador = jogador.split(' ¬∑ ')[0].strip()  # Extrai o nome do jogador
            gols.append(nome_jogador)
    return gols

# Contabilizar gols de jogadores
gols_time_1 = contar_gols(wc['gols_1_detalhes']) + contar_gols(wc['gols_1_penalti'])
gols_time_2 = contar_gols(wc['gols_2_detalhes']) + contar_gols(wc['gols_2_penalti'])

# Usar o Counter para contar as ocorr√™ncias de cada jogador
contador_gols = Counter(gols_time_1 + gols_time_2)

# Converter o contador para um DataFrame
gols_df = pd.DataFrame(contador_gols.items(), columns=['jogador(a)', 'num_gols'])

# Ordenar o DataFrame pelos gols em ordem decrescente
gols_df = gols_df.sort_values(by='num_gols', ascending=False)

# Exibir o top 10 jogadores com mais gols
top_10_jogadores = gols_df.head(10)

# Imprimir a sa√≠da
print(top_10_jogadores)
````
